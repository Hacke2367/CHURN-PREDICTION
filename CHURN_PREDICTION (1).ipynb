# Bivariate Analysis

```python
newdf1_target0 = copydata.loc[copydata['Churn'] == 0]
newdf2_target1 = copydata.loc[copydata['Churn'] == 1]
```

```python
def uniplot(df, col, title, hue=None):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams['axes.labelsize'] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30


    if hue:
        counts = df.groupby([col, hue]).size()
    else:
        counts = df[col].value_counts()
    if (counts == 0).any():
        raise ValueError("Log scale requires all counts to be positive. Found zero counts.")

    fig, ax = plt.subplots()
    hue_vals = df[hue].unique() if hue else []
    n_col = len(df[col].unique())
    n_hue = len(hue_vals) if hue else 1
    width = max(6, (n_col * n_hue) * 0.5 + 2)  # Adjusted width formula
    fig.set_size_inches(width, 8)
    
 
    sns.countplot(
        data=df, 
        x=col, 
        order=df[col].value_counts().index, 
        hue=hue, 
        palette='bright', 
        ax=ax
    )
    

    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
    plt.yscale('log')
    plt.title(title)
    
 
    plt.tight_layout()
    plt.show()
```

```python
uniplot(newdf2_target1, col='Partner', title='Distribution of gender for CHurned Customer', hue='gender')
```
